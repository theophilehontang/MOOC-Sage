```python
def Fibonnaci2(n):
    a=3;b=0;
    c=2;
    for i in range(1,n-1):
        d=a+b
        a=b
        b=c
        c=d
    return(mod(d,n))
```

```python
def un(n):
    a=1
    b=1
    for i in range(1,n):
        c=(a*b)/(3*a+4*b)
        a=b
        b=c
    return(c)
```

```python
def newton(n):
    a=2
    for i in range(1,n+1):
        b=a-f(a)/diff(f)(a)
        a=b
    return(b)
    
```

```python
def halley(n):
    a=2
    for i in range(1,n+1):
        b=a-g(a)/diff(g)(a)
        a=b
    return(b)
```

```python
def baseexo(n,p):
    l=0
    while p^l<=n : 
        l=l+1
    l=l-1
    A=range(l+1)
    a=n%p
    b=((n-a)/p)%p
    A[0]=a
    A[1]=b
    for i in range(1,l):
        a=a+b*p^i
        if i==l:
            b=((n-a)/p^(i+1))
        elif i!=l:
            b=((n-a)/p^(i+1))%p
        A[i+1]=b
    return(A)
```

```python
baseexo(4,2)
```

```python
def legendre2(n,p):
    return((n-sum(baseexo(n,p)))/(p-1))
```
